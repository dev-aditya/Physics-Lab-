```python
import pandas as pd
from numpy import array
import numpy as np
from scipy.stats import linregress
from matplotlib.pyplot import plot
import matplotlib.pyplot as plt
```

```python
fidf = pd.read_csv('tau vs i.csv')
fidf.plot(1,0,kind = 'scatter',marker = '^',color = 'red',label = r'$\tau$ vs $i$')
line1 = linregress(fidf['current'],fidf['tau'])
print(line1)
x = np.linspace(0,4,1000)
plt.plot(x,line1.slope*x + line1.intercept,'-b')
```

```python
fihdf = pd.read_csv("tau vs ih.csv")
fihdf.plot(1,0,kind = 'scatter',marker = '^',color = 'red',label = r'$\tau$ vs $i_h$')
line2 = linregress(fihdf['current(ih)'],fihdf['tau'])
print(line2)
x = np.linspace(0,3,1000)
plt.plot(x,line2.slope*x + line2.intercept,'-b')
```

```python
fthetadf = pd.read_csv("tau vs theta.csv")
fthetadf.plot(1,0,kind = 'scatter',marker = 'x',color = 'green',label = r'$\tau$ vs $sin(\theta)$')
line3 = linregress(fthetadf['sin(theta)'],fthetadf['tau'])
print(line3)
x = np.linspace(1,0,1000)
plt.plot(x,line3.slope*x + line3.intercept,'-b')
plt.legend()
```
